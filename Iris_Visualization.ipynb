● IMPORT IMPORTANT LIBRARIES

```python
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
```

● LOAD IRIS DATASET

```python
df = sns.load_dataset('iris') 
df.head()
```

● DISPLAY DATASET STRUCTURE 

```python
df.shape
```

```python
df.columns
```

SCATTER PLOT b/w SEPAL_LENGTH AND WIDTH

```python
# this will convert the species names to numerical codes for coloring
species = df["species"].astype('category').cat.codes

# Creating the scatter plot
plt.scatter(df["sepal_length"], df["sepal_width"], edgecolors="black", s=88, c=species)

# Customizing the plot with grid
plt.grid(True, linestyle='--', alpha=0.5)

# Setting axis
plt.xlim(4, 8)
plt.ylim(2, 4.5)

# Adding labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Scatter Plot of Sepal Length vs Sepal Width")

# Displaying the plot
plt.show()
```

HISTROGRAM for the DISTRIBUTION_OF_PETAL_LENGTH

```python
# Creating the histogram for petal length
plt.hist(df["petal_length"], bins=20, edgecolor="black")

# Adding title
plt.title("Distribution of Petal Length")

## Setting axis
plt.xlabel("Petal Length")
plt.ylabel("Frequency")

# Adding grid
plt.grid(True, linestyle="--", alpha=0.5)

# Displaying the plot
plt.show()
```

BOXPLOT_OF_PETAL_LENGTH_BY_SPECIES

```python
# Creating the boxplot for petal width by species
sns.boxplot(data=df, x="species", y="petal_width")

# Adding title and showing the plot
plt.title("Boxplot of Petal Length by Species")
plt.show()
```
